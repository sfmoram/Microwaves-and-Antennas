## Transcripci贸n de voz ｏ

Ahora, vamos a transcribir tu archivo de audio. Puedes ajustar la `batch_size` y el `compute_type` seg煤n tus necesidades.

```python
import whisperx
import gc

device = "cuda"
batch_size = 16
compute_type = "float16"

# Transcripci贸n
model = whisperx.load_model("large-v2", device, compute_type=compute_type)
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio, batch_size=batch_size)
print(result["segments"])

# Alineaci贸n
model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)
result = whisperx.align(result["segments"], model_a, metadata, audio, device, return_char_alignments=False)
print(result["segments"])
```

猬锔 Si quieres **descargar la transcripci贸n** como un archivo de texto, ejecuta la siguiente celda.

```python
# Concatenating the transcribed segments
transcription_text = "\n".join([segment['text'] for segment in result["segments"]])

# Writing to a .txt file
with open('transcription.txt', 'w') as file:
    file.write(transcription_text)

# Download the .txt file
files.download('transcription.txt')
```
